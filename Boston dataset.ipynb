# Linear Regression

```python
from sklearn import datasets 
from sklearn import model_selection as ms
from sklearn import linear_model 

data=datasets.load_boston()
data
```

```python
import pandas as pd

x=data.data
y=data.target
col=data.feature_names
df=pd.DataFrame(x,columns=col)
df1=pd.DataFrame(y)
# print(df1)
df
```

```python
# splitting training and testing data
x_train,x_test,y_train,y_test=ms.train_test_split(x,y,random_state=0)

# training algorithm
alg=linear_model.LinearRegression()
alg.fit(x_train,y_train)
m=alg.coef_
c=alg.intercept_

# testing data on algorithm
y_train_predicted=m*x_train +c
y_test_predicted=m*x_test+c

# calculating score
score_train=alg.score(x_train,y_train)
score_test=alg.score(x_test,y_test)
print("m :",m," c :",c)
print("score train :",score_train," score test :",score_test)
```
